```python
import csv

# Classification CSV file name
csv_file = 'aggregated_classifications.csv'

# Initialize an empty list to store the final galaxy IDs
filtered_galaxy_ids = []

uds = []
egs = []

counter_1 = 0
counter_2 = 0

# Read the CSV file
with open(csv_file, mode='r') as file:
    reader = csv.reader(file)
    
    # Skip the header row
    header = next(reader)
    
    # Iterate over each row in the CSV file
    for row_number, row in enumerate(reader, start=2):  
        if len(row) < 11:
            continue
        
        galaxy_id = row[0]         # Column 1: Unique galaxy ID (EGS/UDS)
        class_type = row[9]        # Column 10: Class type 
        confidence = row[10]       # Column 11: Confidence 
        
        # Check if class_type and confidence are not empty
        if not class_type or not confidence:
            continue

        # Strip whitespaces from class_type and confidence
        class_type = class_type.strip()
        confidence = confidence.strip()

        # Check if the class type is 'disks' and confidence is 'True'
        if class_type == 'disk' and confidence == 'True':
            
            # If condition met, extract the last 5 digits of the galaxy ID and append to the list
            filtered_galaxy_ids.append(int(galaxy_id[-5:]))
            
            # Extract EGS galaxies
            if galaxy_id[0] != 'u':
                egs.append(int(galaxy_id[-5:]))
                counter_1 += 1
            
            # Extract UDS galaxies
            if galaxy_id[0] == 'u':
                uds.append(int(galaxy_id[-5:]))
                counter_2 += 1

print("Number of Filtered Disc Galaxy IDs:", len(filtered_galaxy_ids))
print(f'Number of EGS: {counter_1} and Number of UDS {counter_2}')
total = counter_1 + counter_2
print(f'Reconciliation of EGS + UDS: {total}')
```
